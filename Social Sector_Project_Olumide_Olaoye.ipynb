```python
# Convert the Date of Fire to Datetime
data['Date of Fire'] = pd.to_datetime(data['Date of Fire'])
```

`Are there null values in the dataset? If so how many ? and how might you handle them ?`

```python
# Just observing the datasets, I could see some null values, denoted as 'NaN'
# To determine how many null values in a dataset
data.isnull().sum()
```

```python
# To handle null values using fillna() function
data['Identifier'] = data['Identifier'].fillna(0)
data['Src'] = data['Src'].fillna("Not Specified")
data['Type of Market'] = data['Type of Market'].fillna("Not Specified")
data['Actual Start Location'] = data['Actual Start Location'].fillna("Not Specified")
data['Fire put out by'] = data['Fire put out by'].fillna("Not Specified")
data['Reported Start Time'] = data['Reported Start Time'].fillna("Not Specified")
data['Reported Time Put Out'] = data['Reported Time Put Out'].fillna("Not Specified")
data['Fatalities'] = data['Fatalities'].fillna("Not Specified")
data['Looting'] = data['Looting'].fillna("Not Specified")
data['Reported Causes'] = data['Reported Causes'].fillna("Not Specified")
data['Recurring?'] = data['Recurring?'].fillna("Not Specified")
data['When'] = data['When'].fillna("Not Specified")
data['Recurring src'] = data['Recurring src'].fillna("Not Specified")
data['Effective Use of Fire Extinguishers'] = data['Effective Use of Fire Extinguishers'].fillna("Not Specified")
```

`What types of markets are present in the dataset, what is the most common type of market where fire occurs (from the dataset)`

```python
data['Type of Market'].unique()
```

```python
data['Type of Market'].value_counts()
```

```python
plt.title('Market Types with Fire Incidents')
pd.value_counts(data['Type of Market']).plot.bar()
```

The most common type of market where fire occured from the dataset was General Goods, while the least types of markets where fire occured were Roadside Shops, Furniture Market and Electronic Parts

```python
# Create a new column 'Month' from 'Date of Fire'
data['Month'] = data['Date of Fire'].dt.month
```

```python
data.head()
```

```python
plt.title('Fire Incidents by Month')
pd.value_counts(data['Month']).plot.pie()
```

Fire incidents occured more in the month of January, followed by February and occured least in the month of December.

`Visualize the distribution of the counts of fires in each state`

```python
# Checking the value count for Unique states
data['State'].value_counts()
```

```python
plt.title('Fire Incidents by State')
pd.value_counts(data['State']).plot.bar()
```

The state with the highest numbers of fire incidents was Lagos State, as expected, due to the metropolitan nature and population of the state. Ekiti, Abia, Bayelsa, Oyo, FCT, Ogun, and Ebonyi states had the lowest numbers of fire incidents in the year 2020.

`How many fires have been put out by firefighters in the dataset ? How does this number compare with other ways in which fires were put out`

```python
data['Fire put out by'].value_counts()
```

```python
plt.title('Methods of putting out the Fires')
pd.value_counts(data['Fire put out by']).plot.pie()
```

The most commonest ways by which fires were put out was by mostly the Firefighters, while Traders and Army put out the fires sparingly. The other few times fire was put out were not specified in the dataset.

`Visualize the distribution of the counts of reported causes of fire in the dataset`

```python
plt.title('Distribution of reported causes of fire')
pd.value_counts(data['Reported Causes']).plot.pie()
```

The reported causes of fire in these markets in 2020 were mostly unknown from the dataset or they were not reported, while Refuse Fire, Illegal Bush Fire, and Electrical-Power Surge were the least causes of fire incidents.

`Are there recurring fires in the dataset ? If so create a new dataset of recurring fires and extract useful/actionable insights about the markets in which these recurring fires have occured `

```python
data['Recurring?'].value_counts()
```

```python
# Using DataFrame.apply() and lambda function
data["recurring_fires"] = data[["Recurring?", "Fire put out by"]].apply(lambda x: " ".join(x), axis =1)
```

```python
data.head(2)
```

```python
data["recurring_fires"].unique()
```

`Go on from this step to find new insights from the dataset`

```python
plt.title('Markets with recurrence fire incidents')
pd.value_counts(data['recurring_fires']).plot.bar()
```

There were more recurring fire incidents at the markets where there were no firefighters. We had some cases of recurring fire incidents at the marketswhere there were firefighters also. Markets where there were no army and traders had the least recurrence of fire incidents. Other markets were there were small recurrence of fire incidents were not specified in the dataset. There are many factors that could be responsible for the recurrence of fires.
